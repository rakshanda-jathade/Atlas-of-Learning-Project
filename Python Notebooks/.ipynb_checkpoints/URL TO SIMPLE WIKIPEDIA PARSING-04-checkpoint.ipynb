```python
from bs4 import BeautifulSoup as bs

import pandas 


import csv

import requests


import re
```

```python
#df=pd.read_csv("URL1 - Sheet1.csv")
dataset = pandas.read_csv('Biology_URL_Extracted.csv')
dataset.head()
print(dataset)
```

```python
#dataset['Urls'].type()
dtype_before = type(dataset["Urls"]) 
salary_list = dataset["Urls"].tolist()
dtype_after = type(salary_list) 
salary_list 
```

```python
#urls = ['https://simple.wikipedia.org/wiki/Mitochondria','https://simple.wikipedia.org/wiki/Force']
import csv
limit=5

with open('Biology_Context_words_extract.csv','w',encoding='utf-8') as f1:
     writer=csv.writer(f1, delimiter=',',lineterminator='\n',)
    
     for url  in salary_list  :
            zam=url
            f1.write( "%s \n"%(url))
            res = requests.get(url)
            soup = bs(res.text, "html.parser")
            keywords = {}

            for link in soup.find("div",{"id" : "bodyContent"}).findAll("a" , href=re.compile("(/wiki/)")):
                url = link.get("href", " ")
    
                if "/wiki/" in url:
                    keywords[link.text.strip()] = url
             
             
            #f1.write( "%s \n"%(salary_list))
            for key in keywords.keys():
                writer.writerow(( zam,key))
            f1.write("\n")
                    
            print(f1)
            print(url)
            print(keywords)
            print("\n")
           

    

```
